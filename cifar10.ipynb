```python
from keras.datasets import cifar10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train
x_test

x_train=x_train.reshape(50000,32*32*3)
x_train=x_train.astype('float32')/255
x_test=x_test.reshape(10000,32*32*3)
x_test=x_test.astype('float32')/255

from keras.utils import to_categorical
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
model = Sequential()
model.add(Dense(512, activation='relu',input_shape=(32*32*3,)))
model.add(Dense(112,activation='relu'))
model.add(Dense(10, activation='softmax'))


model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()


history = model.fit(x_train, y_train, epochs=500, batch_size=1000,
                    validation_data=(x_test, y_test))
```

```python
model.save('cf10ai.h5')
```

```python
from tensorflow import keras

# Load the saved model
model = keras.models.load_model('cf10ai.h5')

# Check the model summary to verify it loaded correctly
model.summary()
```

```python
from keras.utils import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt

# đặt tên label trong CIFAR-10
cifar10_labels = [
    "Airplane", "Automobile", "Bird", "Cat", "Deer",
    "Dog", "Frog", "Horse", "Ship", "Truck"
]

# Đường dẫn ảnh nhập vào
image_path = "/content/images (3).jpg"

# Tiền xử lý ảnh
img = load_img(image_path, target_size=(32, 32))  # Resize ảnh về 32x32
img_array = img_to_array(img)
img_array = img_array.reshape(1, 32 * 32 * 3)  # Chuyển thành vector 1D
img_array = img_array.astype('float32') / 255.0  # Chuẩn hóa pixel về [0,1]

# Dự đoán class
predicted_index = np.argmax(model.predict(img_array), axis=-1)[0]  # Lấy index class
predicted_label = cifar10_labels[predicted_index]  # Đổi index thành tên class

# In kết quả
print(f"Predicted Class: {predicted_label}")
print(f"Predicted Index: {predicted_index}")

# Hiển thị ảnh gốc + tiêu đề kết quả dự đoán
plt.imshow(load_img(image_path))  # Load ảnh gốc để hiển thị
plt.title(f"Predicted Class: {predicted_label}")  # Gán tiêu đề
plt.axis("off")  # Ẩn trục tọa độ
plt.show()  # Hiển thị ảnh
```
